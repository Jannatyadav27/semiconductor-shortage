```python
data['DATE'] = pd.to_datetime(data['DATE'])
```

```python
print(data.describe())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['DATE'], data['Producer Price Index(By  Industry in $)'], label='Producer Price Index')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Producer Price Index Trends')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['DATE'], data['Producer Price Index(By  Industry in $)'], label='Producer Price Index')
plt.plot(data['DATE'], data['Export Price Index(End Use of semiconductors)'], label='Export Price Index')
plt.plot(data['DATE'], data['Import Price Index(in $ End Use of semiconductors)'], label='Import Price Index')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Price Indices Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['DATE'], data['No. of employed peoples in semiconductor industry'])
plt.xlabel('Date')
plt.ylabel('Number of Employees')
plt.title('Employment Trends in Semiconductor Industry')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['DATE'], data['Relative Importance Weights (Contribution to the Total Industrial Production Index % WISE)'])
plt.xlabel('Date')
plt.ylabel('Relative Importance (%)')
plt.title('Composition of Industrial Production Index')
plt.show()
```
