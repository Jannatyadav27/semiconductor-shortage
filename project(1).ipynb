```python
def calculate_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)  # RMSE
    mae = mean_absolute_error(y_true, y_pred)  # MAE
    r2 = r2_score(y_true, y_pred)  # R-squared
    return rmse, mae, r2
```
