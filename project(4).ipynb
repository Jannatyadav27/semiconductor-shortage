# Forecast of Employment in Semiconductor Industry

```python
from statsmodels.tsa.arima.model import ARIMA
data = pd.read_csv('C:\\Users\\Lenovo\\Downloads\\Semiconductor shortage affects.csv')
```

```python
df = pd.DataFrame(data)
# Converting the "DATE" column to datetime format with the specified format
df["DATE"] = pd.to_datetime(df["DATE"], format="%d-%m-%Y")
df.set_index("DATE", inplace=True)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from datetime import datetime
```

```python
# Define the order and seasonal order for SARIMA model
order = (1, 1, 1) # ARIMA order: (p, d, q)
seasonal_order = (1, 1, 1, 4) # Seasonal order: (P, D, Q, S)
# Create a SARIMAX model with the specified orders and the target variable
model = SARIMAX(df["No. of employed peoples in semiconductor industry"], order=order, seasonal_order=seasonal_order)
# Fiting the SARIMAX model to the data and suppress the display of optimization results
results = model.fit(disp=False)
```

```python
#number of forecast steps
forecast_steps = 5
#forecasted values and confidence intervals using the SARIMA model results
forecast = results.get_forecast(steps=forecast_steps)
forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()
```

```python
print("Forecast for the next few years:")
for DATE, pred, conf_interval in zip(range(2022, 2022 + forecast_steps), forecast_mean, conf_int.values):
    conf_lower, conf_upper = conf_interval
    print(f"year {DATE}: Predicted Employment = {pred:.2f}, Confidence Interval = [{conf_lower:.2f}, {conf_upper:.2f}]")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df["No. of employed peoples in semiconductor industry"], label="Actual Employment")
plt.plot(forecast_mean.index, forecast_mean, color="red", label="Forecasted Employment")
plt.fill_between(conf_int.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color="pink", alpha=0.3, label="Confidence Interval")
plt.xlabel("Year")
plt.ylabel("Employment")
plt.title("Forecast of Employment in Semiconductor Industry")
plt.legend()
plt.grid(True)
plt.show()
```
